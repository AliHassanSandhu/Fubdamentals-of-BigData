```python
%%file Q1(a).py
from mrjob.job import MRJob
import re
class MovieGenreCount(MRJob):
    
    def mapper(self,_,line):
        data = re.split(r',(?=(?:[^"]*"[^"]*")*[^"]*$)', line)
        if data[0]!="":
            year=int(data[0])
        length=data[1]
        genre=data[3]
        if year > 1970 and length!="" and int(length) > 75:
            yield genre,1
    
    def combiner(self, genre, counts):
        yield genre, sum(counts)
        
    def reducer(self, genre, counts):
        yield genre, sum(counts)

if __name__ == '__main__':
    MovieGenreCount.run()
```

```python
%%file Q1(B).py 
import re
from mrjob.job import MRJob
class actress_movie(MRJob):
    def mapper(self,_,line):
        data = re.split(r',(?=(?:[^"]*"[^"]*")*[^"]*$)', line)
        year=data[0]
        movie_name=data[2]
        actress=data[5]
        
        yield actress,(movie_name,year)
        
    def combiner(self,actress,values):
        for val in values:
            if actress=="":
                continue
            else:
                yield actress,(val[0],val[1])
       
    def reducer(self,actress,values):
        for val in values:
            if actress=="":
                continue
            else:
                yield actress,(val[0],val[1])
        
if __name__=='__main__':
    actress_movie.run()
```

```python
%%file Q1(c).py 
import re
from mrjob.job import MRJob
class pop_movies(MRJob):
    def mapper(self, _, line):
        data = re.split(r',(?=(?:[^"]*"[^"]*")*[^"]*$)', line)
        year = int(data[0])
        if year >= 1950 and year <= 1990:
            pop = data[7].strip()
            if pop != '':
                pop = float(pop)
                movie_name = data[2].strip()
                yield None,(pop, movie_name)

    def reducer(self,_,values):
        movies=sorted(values,reverse=True)[:25]
        for  val in movies:
            yield None,(val[1],val[0])
            
    def reducer(self,_,values):
        movies=sorted(values,reverse=True)[:25]
        for  val in movies:
            yield None,(val[1],val[0])
            
            
   
        
if __name__ == '__main__':
    pop_movies.run() 
```

```python
%%file Q1(d).py

from mrjob.job import MRJob
import re

class top5_actor(MRJob):
    
    def mapper(self, _, line):
        
        data = re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        actor=data[4].strip()
        pop=data[6].strip()
        movie_name=data[2].strip()
        
        if actor!='':
            yield actor,(pop,movie_name)
          
    def reducer(self,actor,values):
        movies=sorted(values,reverse=True)[:5]
        mov_str=''
        for val in movies:
            if mov_str=='':
                mov_str=val[0] + "," + val[1]
            else:
                mov_str+="|" + val[0] + "," + val[1]
        yield actor,mov_str        
            
if __name__ == "__main__":
    top5_actor.run()
```

```python
%%file Q1(e).py

from mrjob.job import MRJob
import re

class avg_length(MRJob):
    
    def mapper(self, _, line):
        
        data = re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        genre=data[3].strip()
        length=data[1]
        if genre!="" and length!="":
            yield genre,int(length)
       
    def combiner(self,genre,length):
        count=0
        total=0
        for l in length:
            count=count+1
            total=total+l
            
        yield genre,(total/count)    
        
    def reducer(self,genre,length):
        count=0
        total=0
        for l in length:
            count=count+1
            total=total+l
            
        yield genre,(total/count)      
if __name__=='__main__':
    avg_length.run()
  
```

```python
%%file Q1(f).py
from mrjob.job import MRJob
import re

class lead_actor_actress(MRJob):
    
    def mapper(self, _, line):
        
        data = re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        actor=data[4].strip()
        actress=data[5].strip()
        movie_name=[2]
        if actor!="" and actress!="" and movie_name!="":
            yield (actor,actress),movie_name
            
    def combiner(self,pair,movie_name):
        yield pair,list(movie_name)
        
    def reducer(self,pair,movie_name):
        mov=[]
        for movie in movie_name:
            mov += movie
        if len(set(mov)) > 1:
            yield pair,len(set(mov))
            
if __name__=='__main__':
    lead_actor_actress.run() 
```

```python
%%file Q1(g).py
from mrjob.job import MRJob
import re
class director_genre(MRJob):
    def mapper(self,_,line):
        data = re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        director=data[6].strip()
        genre=data[3].strip()
        if director!="" and genre in ["Mystery","Drama"]:
            yield director,genre
           
    def combiner(self,director,genre):
        yield director,list(genre)
        
    def reducer(self,director,genre):
        gen=[]
        for g in genre:
            gen+=g
        if "Mystery" in gen and "Drama" in gen:
            yield director,len(set(gen))
            
if __name__=='__main__':
     director_genre.run()
        
```

```python
%%file Q1(h).py
from mrjob.job import MRJob
import re
class avg_max_min(MRJob):
    def mapper(self,_,line):
        data = re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        director= data[6].strip()
        popularity=data[7].strip()
        if director!="" and popularity!="":
            yield director,int(popularity)
            
    def reducer(self,director,popularity):
        pop=list(popularity)
        avg=sum(pop)/len(pop)
        maximum=max(pop)
        minimum=min(pop)
        yield director,(avg,maximum,minimum)
if __name__=='__main__':
    avg_max_min.run()
```

```python
%%file Q2.py 
from mrjob.job import MRJob
import re
class SimilarMovies(MRJob):
    
    def mapper(self, _, line):
        
        data= re.split(r',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', line)
        year = int(data[0].strip())
        if year < 1970:
            title = data[2].strip()
            genre = data[3].strip()
            director = data[6].strip()
            actor = data[4].strip()
            actress = data[5].strip()
            key = (genre, director, actor, actress)
            
            if "" not in key:
                yield key, title
            
    def reducer(self, key, values):
        
        movie_list = list(values)
        if len(movie_list) > 1:
            for movie in movie_list:
                yield movie, key
                
if __name__ == "__main__":
    SimilarMovies.run()               
```
